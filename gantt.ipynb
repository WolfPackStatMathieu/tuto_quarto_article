```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
from datetime import datetime


# Prparer les donnes
data = {
    'Tche': ['Introduction  la cyberscurit et\n renforcement des comptences en ML/DL'
    ,'Analyse des besoins, dfinition des objectifs,\n identification des sources de donnes', 
    'Collecte et prparation des jeux de donnes simuls,\n analyse exploratoire des donnes.'
    , 'Prototypage et\n exprimentations avec des algorithmes de dtection'
    , ' Documentation des rsultats (note intermdiaire),\n planification des prochaines tapes'
    , 'Accs aux donnes relles, prparation \n et feature engineering.'
    , 'Entranement des modles, valuation \net optimisation.'
    ,'Dveloppement de l\'outil \n(backend et frontend).'
    , 'Intgration avec le SIEM, automatisation \net dploiement en production.'
    , ' Tests et validation.'
    , 'Documentation et formation \ndes utilisateurs.'
    ], 
    'Date de dbut': ['2024-04-01', '2024-04-15', '2024-04-29'
                    , '2024-05-20', '2024-06-03', '2024-06-17'
                    , '2024-07-01', '2024-07-15', '2024-07-29'
                    , '2024-08-12', '2024-08-19'
                    ],
    'Dure (jours)': [14, 14, 21, 14, 14, 14, 14, 14, 14, 7, 7]
}

# Convertir les donnes en DataFrame
df = pd.DataFrame(data)
df['Date de dbut'] = pd.to_datetime(df['Date de dbut'])
df['Date de fin'] = df['Date de dbut'] + pd.to_timedelta(df['Dure (jours)'], unit='d')

# Crer le diagramme de Gantt
fig, ax = plt.subplots(figsize=(10, 6))

for i, (task, start, end) in enumerate(zip(df['Tche'], df['Date de dbut'], df['Date de fin'])):
    ax.barh(task, (end - start).days, left=start, height=0.4)

# Configurer les axes
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)
plt.xlabel('Dates')
plt.ylabel('Tches')
plt.title('Diagramme de Gantt')

# Afficher le diagramme
plt.tight_layout()
plt.show()
```
