```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
from datetime import datetime


# Préparer les données
data = {
    'Tâche': ['Collecte de données', 'Préparation des données', 'Analyse exploratoire', 'Modélisation', 'Évaluation des modèles', 'Rédaction du rapport'],
    'Date de début': ['2024-06-01', '2024-06-08', '2024-06-13', '2024-06-23', '2024-07-08', '2024-07-15'],
    'Durée (jours)': [7, 5, 10, 15, 7, 10]
}

# Convertir les données en DataFrame
df = pd.DataFrame(data)
df['Date de début'] = pd.to_datetime(df['Date de début'])
df['Date de fin'] = df['Date de début'] + pd.to_timedelta(df['Durée (jours)'], unit='d')

# Créer le diagramme de Gantt
fig, ax = plt.subplots(figsize=(10, 6))

for i, (task, start, end) in enumerate(zip(df['Tâche'], df['Date de début'], df['Date de fin'])):
    ax.barh(task, (end - start).days, left=start, height=0.4)

# Configurer les axes
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)
plt.xlabel('Dates')
plt.ylabel('Tâches')
plt.title('Diagramme de Gantt')

# Afficher le diagramme
plt.tight_layout()
plt.show()
```
